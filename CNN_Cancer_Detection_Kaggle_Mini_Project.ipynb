# CNN Cancer Detection Kaggle Mini-Project

# 1.Brief description of the problem and data

The data set contains about 220k tissue images. Some of them contains cancer cells in the middle of the image, while the others don’t. The goal of this problem is to predict whether a given tissue image has cancer cells in the center field.

# 2.Exploratory Data Analysis (EDA) — Inspect, Visualize and Clean the Data

# Model Architecture

The network consists of several layers arranged in a sequential order within its features attribute. Initially, it starts with a convolutional layer that takes an input with 3 channels (e.g., RGB image) and outputs 64 feature maps, followed by batch normalization and a ReLU activation function, and a max pooling layer to reduce spatial dimensions. This pattern is repeated three more times with increasing number of filters (64 to 128, then 128 to 256, and finally 256 to 1024), each time followed by batch normalization, ReLU activation, and max pooling.After passing through these convolutional blocks, the output is processed by an adaptive average pooling layer, reducing the feature map to a size of 1x1. The network's classifier part consists of a fully connected layer that reduces the dimension from 1024 to 64, applies ReLU activation and dropout for regularization, and finally outputs to a layer with 2 units.

# Results and Analysis

I have reached 91% acc using a simple CNN model. While I have tried some of the traditional ML methods like SVM to solve the problem, the results are bad. I have also tried tuning the model size, increasing CNN layers or channel number. The run time significantly increased and I have to reduce batch size, but the accuracy merely increased by 0.01%

# Conclusion

Moving from SVM to CNN network significantly improved the result. This shows the power of CNN on 2D image inputs. I have also learned larger models are slower for sure, but not necessarily better.
